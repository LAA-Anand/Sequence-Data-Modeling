```python
#Usage of TFIDF Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
import pandas as pd
```

```python
ds = pd.read_csv("/content/drive/My Drive/Recommendation System Mahesh Anand/fashion_store.csv")
```

```python
tf = TfidfVectorizer(analyzer='word', ngram_range=(1, 3), min_df=0.0, stop_words='english')
#min_df is used for removing terms that appear too infrequently
#min_df = 0.01 means "ignore terms that appear in less than 1% of the documents
tfidf_matrix = tf.fit_transform(ds['description'])
```

```python
tfidf_matrix.shape
```

```python
type(tfidf_matrix)
```

```python
cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
cosine_similarities.shape
```

```python
cosine_similarities
```

```python
cosine_similarities[0]
```

```python
#by default smallest to largest
cosine_similarities[0].argsort()
```

```python
similar_indices = cosine_similarities[200].argsort()[-6:-1]
```

```python
from sklearn.neighbors import NearestNeighbors
model_knn = NearestNeighbors(metric = 'cosine')
model_knn.fit(tfidf_matrix)
```

```python
distances, indices = model_knn.kneighbors(tfidf_matrix[200, :], n_neighbors = 6)
```
