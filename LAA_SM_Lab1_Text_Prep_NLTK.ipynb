Examples of English stop words often include:

    Articles: "a," "an," "the"
    Conjunctions: "and," "but," "if"
    Prepositions: "at," "by," "into," "on"
    Pronouns: "he," "she," "it," "they"
    Auxiliary Verbs: "am," "is," "are," "was," "were"
    Others: "do," "so," "such," "very"

```python
# Step 4: Removing Stopwords
stop_words = set(stopwords.words('english'))
words_no_stop = [word for word in words_no_punc if word not in stop_words]
```

The PorterStemmer is an algorithm used for stemming in natural language processing, which involves reducing words to their base or root form. It simplifies words to their root forms. For example, the words "running," "runner," and "ran" might all be reduced to "run."

```python
# Step 5: Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in words_no_stop]
```

In modern day LLMs, the most common method is subword tokenization, which breaks words down into meaningful subunits (subwords or word pieces). This approach captures the morphology of the language better than stemming, as it allows the model to handle unknown words more gracefully by breaking them down into known subunits (Byte-Pair Encoding)

```python
print(stemmed_words)
```

```python
len(stemmed_words)
```

```python
# Step 6: Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in words_no_stop]
```

WordNetLemmatizer uses the WordNet lexical database, which groups English words into sets of synonyms (synsets) and records a range of relations among these synsets or their members.


It can consider the part of speech (POS) of a word, which allows it to distinguish between meanings and ensure that the lemmatization is contextually appropriate. For example, the word "better" can be an adjective or a verb, and knowing the correct POS helps the lemmatizer return the most accurate lemma ("good" as an adjective and "better" as a verb).

```python
print(lemmatized_words)
```

```python
# Example words
print("Lemmatization of 'better', as an adjective:", lemmatizer.lemmatize('better', pos='a'))  # good
print("Lemmatization of 'running', as a verb:", lemmatizer.lemmatize('running', pos='v'))  # run
```

```python
# Calculate frequency distribution of the words
freq_dist = FreqDist(hindi_words)

# Display the 10 most common words
print("Top 10 most common words in the Hindi corpus:")
for word, frequency in freq_dist.most_common(10):
    print(f"{word}: {frequency}")
```
