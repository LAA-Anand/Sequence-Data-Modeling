```python
text_clean_sw = [word for word in text_clean if word not in stopwords]
print(text_clean_sw)
```

Common POS Tags in spaCy

    ADJ: adjective (e.g., big, old)
    ADV: adverb (e.g., very, well)
    INTJ: interjection (e.g., psst, ouch)
    NOUN: noun (e.g., dog, chair)
    PROPN: proper noun (e.g., Mary, John)
    VERB: verb (e.g., run, eats)
    ADP: adposition (prepositions and postpositions; e.g., in, to, during)
    AUX: auxiliary verb (e.g., has (in "has gone"), is (in "is running"))
    CONJ: coordinating conjunction (e.g., and, or)
    DET: determiner (e.g., a, an, the)
    NUM: numeral (e.g., one, 2017)
    PART: particle (e.g., 's, not)
    PRON: pronoun (e.g., she, they)
    SCONJ: subordinating conjunction (e.g., although, because)
    PUNCT: punctuation (e.g., ., (, ))
    SYM: symbol (e.g., $, %, @)
    X: other (e.g., sfpksdpsxms)
    SPACE: space (e.g., \n, \t, spaces)

```python
pos_tags = [(token.text, token.pos_) for token in doc]
print(pos_tags)
```

```python
# Process some text
ex1 = "Walmart is looking at buying U.K. startup for $1 billion"
text1 = nlp(ex1)
pos_tags = [(token.text, token.pos_,spacy.explain(token.pos_)) for token in text1]
print(pos_tags)
```

```python
#Named Entity Recognition is the process of finding named entities from the text and
#assigning them to categories such as organization’s names, timelines, dates, etc.
ex2="Sundar Pichai, the CEO of Alphabet Inc. is walking in the streets of California. "
text2 = nlp(ex2)
for ent in text2.ents:
    print(ent.text,ent.label_, spacy.explain(ent.label_))
```

```python
ex3="John, from Tiger Analytics is walking in the streets of Bangalore. "
text3 = nlp(ex3)
for ent in text3.ents:
    print(ent.text,ent.label_, spacy.explain(ent.label_))
```
